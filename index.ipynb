```python
# Let's get started importing the necessary libraries
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
# Loading the data
df = pd.read_csv('ames_train.csv')
```

```python
# Investigate the Data
df.info()
```

```python
df.describe()
```

```python
# Investigating Distributions using scatter_matrix
df.columns = [col.lower() for col in df.columns]
df.columns
#which columns am I interested in? 
#sales price, year sold, year built, year remodeled, home style, street
```

```python
# Create a plot that shows the SalesPrice Distribution

new_df = df[['street', 'yearbuilt', 'yearremodadd', 'housestyle', 'yrsold' ,'saleprice']]
axes = pd.plotting.scatter_matrix(new_df, figsize=(10,10))
plt.show()
```

```python
# Create a plot that shows the LotArea Distribution

fig, ax = plt.subplots()
ax.hist(df['saleprice'], bins='auto')
ax.set_title('Distribution of Sale Prices')
ax.set_xlabel('Sale Price')
ax.set_ylabel('Number of houses');
```

```python
# Create a plot that shows the Distribution of the overall house condition

fig, ax = plt.subplots()
ax.hist(df['overallcond'], bins='auto')
ax.set_title('Condition of Houses')
ax.set_xlabel('Condition')
ax.set_ylabel('Number of houses');
```

```python
# Create a Box Plot for SalePrice

fig, ax = plt.subplots()
ax.boxplot(df['saleprice'])
ax.set_title('Sales Price')
```

```python
# Perform an Exploration of home values by age

df['age'] = df['yrsold'] - df['yearbuilt']

df.plot.scatter('saleprice', 'age')

#the newer the home the greater the price
```
